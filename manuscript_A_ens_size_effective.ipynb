```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(2004, 2024)
years_hist = np.arange(1940, 2004)

ens_f = np.ones(len(years)) * 20
ens_cf = np.ones(len(years)) * 20
ens_hist = np.ones(len(years_hist)) * 10

# Missing members
ens_f[2005-2004] -= 1
ens_f[2012-2004] -= 1
ens_f[2017-2004] -= 3
ens_f[2021-2004] -= 1

ens_cf[2009-2004] -= 1
ens_cf[2017-2004] -= 1

ens_hist[1992-1940] -= 1

fig, ax = plt.subplots(figsize=(10,5))

width = 0.4

ax.bar(
    years_hist,
    ens_hist,
    width=width,
    color='tab:orange',
    label='Factual'
)

ax.bar(
    years - width/2,
    ens_f,
    width=width,
    color='tab:orange'
)

ax.bar(
    years + width/2,
    ens_cf,
    width=width,
    color='tab:green',
    label='Counterfactual'
)

ax.set_xlabel("Year")
ax.set_ylabel("Number of ensemble members")
ax.set_title("Effective ensemble size by year")

all_years = np.concatenate((years_hist, years))
ax.set_xticks(all_years[::5])
ax.set_yticks(np.arange(0,20+2,2))
ax.legend()
ax.set_ylim(0, 22)
ax.set_xlim(1939,2024)

plt.tight_layout()
plt.savefig("./figs/SI_fig_ens_size_effective.png", dpi=300)
plt.show()
```
